```python
#Magic 8-Ball Assignment:

# Create definition mimicking Magic 8-Ball that randomly selects and displays a response from 
# the Answers list to an inputted question

def Magic8Ball():

# import random used later for selecting from magicBallAnswers variable list

    import random

# List of responses from which one reply will be randomly selected and displayed to a question

    magicBallAnswers = [
    "It is certain",
    "It is decidely so",
    "Without a doubt",
    "Yes definitely",
    "You may rely on it",
    "As I see it, yes",
    "Most Likely",
    "Outlook good",
    "Yes",
    "Signs point to yes",
    "Reply hazy try again",
    "Ask again later",
    "Better not tell you now",
    "Cannot predict now",
    "Concentrate and try again",
    "Don't count on it",
    "My reply is no",
    "My sources say no",
    "Outlook not so good",
    "Very doubtful"
    ]

# Time for "Magic 8Ball" to prompt for question and provide answer!

    question = "y"
    
    while question != "n":
        
        question = input("The Magic 8-Ball knows all!  What is your burning question? ==> ")
        print(" ")
        
        print("YOU ASKED: ",question)
        print("MAGIC 8-BALL SAYS: ",random.choice(magicBallAnswers))
        
        print(" ")
        question = input("Do you have another question for the Magic 8-ball? (y/n)")
  
    print(" ")
    print("Thank you for playing!")

    
Magic8Ball()
```
